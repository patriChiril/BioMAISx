```python
from evaluate_annotation_set import open_and_format_annotation_results, create_comparison_table
from pathlib import Path
import pandas as pd
import numpy as np

root = Path("").resolve().parent
```

```python
# compare accross multiple batches
results_directory = root / "annotation" / "cikm"
# will compare all batches starting with [batch_set], leave empty string to compare two people accross all
batch_set = "mbio-factiva-en-lrec"


person_a = "patricia"
person_b = "trevor"

person_a_files = [file for file in results_directory.iterdir() if file.stem.startswith(batch_set) and file.stem.endswith(person_a)]
person_b_files = [file for file in results_directory.iterdir() if file.stem.startswith(batch_set) and file.stem.endswith(person_b)]

print(f"All annotations matching: {batch_set} - {person_a}: {len(person_a_files)}, {person_b}: {len(person_b_files)}")

person_a_dfs = [open_and_format_annotation_results(person_a_file) for person_a_file in person_a_files]
person_b_dfs = [open_and_format_annotation_results(person_b_file) for person_b_file in person_b_files]

results_person_a = pd.concat(person_a_dfs)
results_person_b = pd.concat(person_b_dfs)

comparison_table = create_comparison_table(results_person_a, results_person_b)
comparison_table = comparison_table.fillna("NA")

```

```python
# pick one of 'entity_type', 'aspect', 'sentiment' to see confusion matrix of
ELEMENT = 'aspect'
pd.crosstab(comparison_table[f"{ELEMENT}_a"], comparison_table[f"{ELEMENT}_b"], dropna=False)
```
